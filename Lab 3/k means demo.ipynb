```python
from sklearn import datasets
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
```

```python
iris = datasets.load_iris()
```

```python
import pandas as pd
X=pd.read_csv("C:\\Users\\Rupali\\Desktop\\Machine Learning\\Odd2020\\BCA\\6 BCA 2020\\glass.csv")
```

```python
X
```

```python
y=X['Type']
X=X.drop(['Type'], axis=1)
```

```python
X = iris.data[:, :2]
y = iris.target
```

```python
plt.scatter(X.iloc[:,0], X.iloc[:,1], c=y, cmap='gist_rainbow')
plt.xlabel('Ri', fontsize=18)
plt.ylabel('Na', fontsize=18)
```

```python
km = KMeans(n_clusters = 7,random_state=21)
km.fit(X)
```

```python
centers = km.cluster_centers_
print(centers)
```

```python
km.labels_
```

```python
new_labels = km.labels_
# Plot the identified clusters and compare with the answers
fig, axes = plt.subplots(1, 2, figsize=(16,8))
axes[0].scatter(X.iloc[:, 0], X.iloc[:, 1], c=y, cmap='gist_rainbow',
edgecolor='k', s=150)
axes[1].scatter(X.iloc[:, 0], X.iloc[:, 1], c=new_labels, cmap='gist_rainbow',
edgecolor='k', s=150)
axes[0].set_xlabel('RI', fontsize=18)
axes[0].set_ylabel('NA', fontsize=18)
axes[1].set_xlabel('RI', fontsize=18)
axes[1].set_ylabel('NA', fontsize=18)
axes[0].tick_params(direction='in', length=10, width=5, colors='k', labelsize=20)
axes[1].tick_params(direction='in', length=10, width=5, colors='k', labelsize=20)
axes[0].set_title('Actual', fontsize=18)
axes[1].set_title('Predicted', fontsize=18)
```

```python
km.inertia_
```

```python
distortions = []
K = range(1,10)
for k in K:
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(X)
    distortions.append(kmeanModel.inertia_)
```

```python
plt.figure(figsize=(16,8))
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('WCSS')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```
